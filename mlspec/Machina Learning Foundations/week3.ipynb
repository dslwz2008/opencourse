```python
import graphlab
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products = graphlab.SFrame('amazon_baby.gl/')
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
products.head()
```

# Problem1

```python
products['awesome'] = products['word_count'].apply(lambda x:x.get('awesome',0))
```

```python
for word in selected_words:
    products[word] = products['word_count'].apply(lambda x:x.get(word,0))
```

```python
products.head()
```

```python
import operator
dictCount= {}
for word in selected_words:
    totalCount = products[word].sum()
    dictCount[word] = totalCount
sortedDict = sorted(dictCount.iteritems(),key=operator.itemgetter(1))
print(dictCount)
print(sortedDict)
```

# Problem 2

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
#ignore all 3* reviews
products = products[products['rating'] != 3]
#positive sentiment = 4* or 5* reviews
products['sentiment'] = products['rating'] >=4
train_data,test_data = products.random_split(.8, seed=0)
selected_words_model=graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=selected_words,
                                                     validation_set=test_data)
```

```python
coef = selected_words_model['coefficients']
coef.sort('value')
```

# Problem3

```python
selected_words_model.evaluate(test_data)
```
